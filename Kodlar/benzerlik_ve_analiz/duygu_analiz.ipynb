```python
from transformers import BertTokenizer, TFBertForSequenceClassification
import tensorflow as tf

# Dosyadan metinleri okuma
file_path = "ilk_100_cumle.txt"
with open(file_path, "r", encoding="utf-8") as file:
    texts = file.readlines()

# Önceden eğitilmiş BERT modelini ve tokenizer'ını yükleyelim
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = TFBertForSequenceClassification.from_pretrained(model_name)

# Metinleri BERT modeline uygun hale getirme
input_ids = tokenizer(texts, return_tensors="tf", padding=True, truncation=True)["input_ids"]

# Metinleri modelde değerlendirme
outputs = model(input_ids)
predictions = tf.nn.softmax(outputs.logits, axis=-1)

# Sentiment etiketlerini ve güven skorlarını yazdırma
for text, prediction in zip(texts, predictions):
    label = tf.argmax(prediction).numpy()
    confidence = prediction[label].numpy()
    print(f"Text: {text.strip()}\nSentiment: {label}, Confidence: {confidence}\n")
```
